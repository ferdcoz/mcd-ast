# **WAPP**

```python
wapp_pred = pd.read_excel('C:/Users/jnnov/OneDrive/Documentos/AUSTRAL/MATERIAS/1_CUAT_2DO/SERIES_TEMPORALES/TP 2/WAPP_PRED.xlsx')  # Reemplaza 'ruta/del/archivo' con la ubicación real del archivo

# Mostrar las primeras filas para verificar la carga exitosa
print(wapp_pred.head())
```

```python
import matplotlib.pyplot as plt


# Convertir la columna FECHA a tipo datetime si aún no está en ese formato
wapp_pred['FECHA'] = pd.to_datetime(wapp_pred['FECHA'])

# Establecer la columna FECHA como el índice del DataFrame (opcional, pero puede ser útil)
wapp_pred.set_index('FECHA', inplace=True)
```

```python

# Graficar las series con colores diferentes
plt.figure(figsize=(20, 6))  # Tamaño de la figura
#colors = ['#8bc34a', '#009688', '#03a9f4', '#3f51b5', '#673ab7', '#9c27b0', '#9467bd'] #'ffeb3b


# Graficar cada serie con un color distinto
plt.plot(wapp_pred['WAPP_XGB'], label='WAPP_XGB', color='#673ab7', linewidth=1, alpha=0.8) #925C91
plt.plot(wapp_pred['WAPP_LGBM'], label='WAPP_LGBM', color='#3f51b5', linewidth=1, alpha=0.8) #EFE978
plt.plot(wapp_pred['WAPP_AUTOML'], label='WAPP_AUTOML', color='#03a9f4', linewidth=1, alpha=0.8) #58508d
plt.plot(wapp_pred['WAPP_NEURAL_PROPHET'], label='WAPP_NEURAL_PROPHET', color='#009688', linewidth=1, alpha=0.8)
plt.plot(wapp_pred['WAPP_PROPHET'], label='WAPP_PROPHET', color='#8bc34a', linewidth=1, alpha=0.8)
plt.plot(wapp_pred['WAPP_LSTM'], label='WAPP_LSTM', color='#ffc107', linewidth=1, alpha=0.8)

plt.plot(wapp_pred['WAPP_REAL'], label='WAPP_REAL', color='#f44336')
ticks = wapp_pred.index[::5]
plt.xticks(ticks, rotation=90)
# Agregar etiquetas y leyenda
plt.xlabel('Fecha')
plt.ylabel('Valores')
plt.title('Predicciones mensajes de Whatsapp (WAPP)')
plt.legend()


# Mostrar el gráfico
plt.show()
```
